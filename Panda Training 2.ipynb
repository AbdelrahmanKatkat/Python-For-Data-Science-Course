## Missing Data (dropna,fillna)

```python
#Panda will fill missing data with NaN
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
d={'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]}
```

```python
df=pd.DataFrame(d)
```

```python
 df
```

```python
# drop missing data (dropna() will delete any row, or colum with missin values)
```

```python
df.dropna() # drop rows
```

```python
df.dropna(axis=1) 
    # change axis to drop columns
```

```python
# Threshold with dropna to choose  min non NaN values 
```

```python
df.dropna(thresh=2) 
    # here non NaN is 2 ( if there is two(thresh hold number) existing values keep the row)
```

```python
df.fillna(value='Fill Value')
```

```python
df.fillna(value=df.mean())# it filled missing values with mean
```

## Groupby

```python
# It allows you to group together rows based off of a column and perform an aggregate(takes many values and produce only one value) function on them 
```

```python
import numpy  as np 
import pandas as pd
```

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Company')
    #that returns a grouby object and it's memory locattion 
```

```python
byComp=df.groupby('Company') 
    # We can perform a fn for an object now
```

```python
byComp.mean() 
    # Pandas will automatilcally ignore any non numerical column like  person name
```

```python
byComp.sum()
```

```python
byComp.std()
```

```python
byComp.sum().loc['FB']
    # The result is a data frame, so I can perform any calling method on it 
```

```python
df.groupby('Company').count() # Here it returned person as it's a number
```

```python
df.groupby('Company').max()
```

```python
df.groupby('Company').describe().transpose()
    # Transpose will rotate my elements
```

```python
df.groupby('Company').describe().transpose()['FB']
```

```python
df.groupby('Company').describe().loc['FB']
```

## Merging Joining and Concatenating

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
#Concatenating means to merge data, but data on merging axis should be similar 
```

```python
pd.concat([df1,df2,df3]) # axis =0 as default on x axis
```

```python
pd.concat([df1,df2,df3],1) # here many missing data, so have enough information on concatenate on 
```

```python
# Merging

# To merge logic as in sequal merging 
# here you will path on a single column key not an index
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})    
```

```python
left 
```

```python
right
```

```python
pd.merge(left,right,how='inner')
```

```python
pd.merge(left,right,how='inner',on='key') # that will merge on key 
```

```python
pd.merge(left,right,on=['key1','key2']) # puts different keys to merge on
```

```python
pd.merge(left,right,how='outer',on=['key1','key2'])
```

```python
pd.merge(left,right)
```

```python
pd.merge(left,right,how='left')
```

```python
# Joining
# Used to combine columns of differnt indexed dataframes 
# It's like merge, but keys on index not column
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left.join(right)# default is inner
```

```python
left
```

```python
right
```

```python
left.join(right,how='outer')
```
