## Built in viz

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
```

```python
df1=pd.read_csv('df1',index_col=0)
    # you can identify which col to be index
```

```python
df1
    # index is time series here
```

```python
df2=pd.read_csv('df2')
```

```python
df2
    # index here is just sequential index
```

```python
# Built in visualization
    # You can use two methods:
```

```python
    # Method 1
df1['A'].plot.hist(bins=50)
    # pandas can call histplot on it's own
    # Argument can be added to
```

```python
    # Method 2
df1['A'].plot(kind='hist')
```

```python
df2['a'].plot.bar()
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
    # To make data stacked over each other
```

```python
# line plot
    # Relation between two axis x,y
```

```python
df1.head()
```

```python
df1['A'].plot.line(x=df1.index)
```

```python
df1['B'].plot.line(x=df1.index,figsize=(12,3),lw=2,ls='--')
    # As it follows matplot you can use figsize
```

```python
# scatter plot(x,y)
```

```python
df1.plot.scatter(x='A',y='B',c='r')
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
    # you can define the color based on other column
    # It's a 3D lvl
```

```python
df1.head()
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*99)
    # You can show another column by size 
```

```python
# box plot
```

```python
df2.plot.box()
    # column will be categories
    # index will be y axis
```

```python
df2[['a','b']].plot.box()
```

```python
# hexbin
    # It's like scatter plot but with hexs
```

```python
df=pd.DataFrame(np.random.rand(1000,2),columns=['a','b'])
```

```python
df.head()
```

```python
df.plot.hexbin(x='a',y='b')
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm')
    # To increase the hex size for grids
    # plots get darkes as you find more data in it 
```
