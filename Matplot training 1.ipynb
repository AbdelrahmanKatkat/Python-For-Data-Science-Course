# Matplotlib with python

```python
 import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# To see the plot in jupyter
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
# Functional methond to make plot
```

```python
plt.plot(x,y)
#plt.show () in other IDE (it's like print)
plt.xlabel('X Label')
# To name the x axis
plt.ylabel('Y Label')
# To name the y axis
plt.title('Title')
# To name the title
```

```python
plt.plot(x,y,'r')
# To change color
```

```python
# subplot
plt.subplot(1,2,1)
# First, rows
# Second, columns
# Third, plot number
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented method to make plot: 
    # The main idea to make a figure object an empty canvas
    # Then, you call methods to it like axes of canvas in relation to canvas
    # Then, you call methods to methods it self
```

```python
fig=plt.figure() 
    # to make a figure object( an imaginary blank canvas), and we can axis to that campus
axes=fig.add_axes([.1,.1,.8,.8])
    # Here we called the method object add axes
    # First parameter, the left of axes
    # Second parameter, the bottom of axes
    # Third, the width
    # Forth, the Height
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
    # all of that is other methods for figure:
plt.figure().add_axes([.1,.1,.8,.8]).plot(x,y)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,.1,.8,.8])
axes2=fig.add_axes([.2,.5,.4,.3])
    # All the values is a percent from 0 to 1
    # All the values is percent from the blank canvas it self 
    # axes2 = 20% from left, 50% form bottom, and so on.
axes1.plot(x,y)
axes2.plot(y,x)
axes2.set_title('Smaller plot')
axes1.set_title('Bigger plot')
axes2.set_xlabel('Xlabel')
axes2.set_ylabel('Ylabel')
```

```python
# subplot in object oriented method is another method added to figure
# subplot will set the axes automatically for you
fig, axes= plt.subplots(nrows=1,ncols=1)
    # That line is easier, as you assigned both axes, figure.Then, you plotted
    # We need to open use that line every time to plot fig
```

```python
fig=plt.figure() # To plot
sub=fig.subplots(3,3)
    # Here I'm showing that it's just another method of figure
plt.tight_layout() 
    # To fix the over lapping 
```

```python
fig.subplots(3,3)
    # Here I can't plot, as I didn't use plt to make a figure
```

```python
fig, axes= plt.subplots(1,2) 
    # All orders in one(Tuple unpacking)
```

```python
axes 
# assigned axes here is an array, so I can ilterate through it 
```

```python
fig, axes=plt.subplots(1,2)
for ax in axes:
    ax.plot(x,y)
    # We plotted through axes, as it's an array
   
```

```python
# As it's array, we can:
fig, axes=plt.subplots(1,2) 
axes[0].plot(x,y)
axes[0].set_title('1st')
axes[1].plot(y,x)
axes[1].set_title('2nd')
```

## Figure size, aspect ratio, and DPI

```python
# Matplot allows you to control each of them when you call figure object
    #DPI: dot per inch, or pixels per inch
    #Figure size is a Tuple
    #
```

```python
fig=plt.figure(figsize=(8,2),dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# To control figsize with subplot
fig, axes= plt.subplots(1,2,figsize=(10,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout
```

## legend

```python
# legend fn searches in your plot,and try to find if you labled the plot
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)

ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label= 'X Cubed')
    # Here we labeled before using legend()
ax.legend(loc=(1,1))
    # You can define legend location loc=0,1,2,....
    # Use loc=0 matplot will find best location for you
    # You can define location by tuple loc=(x,y)
    # Note (x,y) in tuple is percent form 0 to 1
```

## Customizing your graph

```python
# Setting Colors
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange')
ax.plot(y,x,'#FFBC00')
    # It can take many argument: 
    # Colors name as strings, #Argb hex code,
```

```python
# Line Width and Style
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,linewidth=2,alpha=.5)
    # Default linedwidth is 1
    # Alpha argument for transperancy
    # you can write just lw
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,linewidth=2,alpha=.5,linestyle='--')
    # Style: means the type of your line
    # Examples: '--','.','.-','steps'
    # You can just write ls
```

```python
# Markers
    # Used when you have a small data points
```

```python
x
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,linewidth=1,alpha=.5,linestyle='--',marker='o',
         markersize=2, markerfacecolor='yellow', markeredgewidth=3)
    # It will mark each data point
    # E.g.:'+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
    # markersize increase width, and default it 1
    # markerfacecolor
    # markeredgecolor
    # markeredgewidth
    
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,linewidth=.5,alpha=.5,linestyle='--',marker='s',
         markersize=8
        , markerfacecolor='yellow', markeredgewidth=2)
```

```python
# Axis apperance & plot Range
    # Plot range: How much to show
    
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x, y, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
ax.set_xlim([2, 10])
ax.set_ylim([0,10])
    # Here you will define the upper and lower bound of tuples
    # Note that if you don't have enough points the plot will be lines
```
