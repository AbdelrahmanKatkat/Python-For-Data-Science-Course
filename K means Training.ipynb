<font color='red' size='5px'> K-means Theory <font/>
    


- It is an unsupervised learning algorithm that attempt to group similar clusters together in your data
    
- Examples:
    - Cluster similar documets 
    - cluster customers based on features
    - Market segmentation
    - identify similar physical groups
    
- Algorithm:
    - Choose number of Clusters "K"
    - Randomly assign a centorid place 
        - Better to take that random place as an existing point.
    - Randomly assign each point to the nearest cluster.
    - Enter the following loop till Cluster location stop changing.
        - For each cluster, compute the cluster centroid by taking the mean vector of point in the cluster
        - Assign each data point to the cluster for which centroid is closest "Minimize the distance".

- Choosing a K Value 
    - Elbow method
        - Calculate Sum of Squared Error for each cluster (SSE). Also called Distortion function.
            - ex: SSE for K=2,4,6,8, etc.
        - Plot K against the SSE
        - Choose the k the produce the elbow effect, "SSE decreases abruptly".
    - Downstream purpose
        - From your experience you know the end value, so choose the K depending on it.
            - EX: in Closthes store sizes are small, medium, and large, So K=3.mm
